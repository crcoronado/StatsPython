## Loading Datasets
It si recommended to use *pandas.read_csv* command, but other options are:

*Manual Load*

*numpy.loadtxt*

*numpy.genfromtxt*

*pickle*

**Examples**

```python
import numpy as np
import pickle
import pandas as pd
filename = 'load.csv'
```

```python
#Manual Load
cols = None
data = []
with open(filename) as f:
    for line in f.readlines():
        vals = line.replace("\n", "").split(",")
        if cols is None:
            cols = vals
        else:
            data.append([float(x) for x in vals])

d0 =  pd.DataFrame(data, columns=cols)
print(d0.dtypes)
d0.head()
                
```

```python
#np.loadtxt
d1 = np.loadtxt(filename, skiprows=1, delimiter=",")
print(d1.dtype)
print(d1[:5, :])
```

```python
#np.genfromtxt
d2 = np.genfromtxt(filename, delimiter=",", names=True, dtype=None)
print(d2.dtype)
print(d2[:5])
```

```python
#pd.read_csv
d3 = pd.read_csv(filename)
print(d3.dtypes)
d3.head()
```

```python
#With Pickle
with open("load_pickle.pickle", "rb") as f:
    d4 = pickle.load(f)
print(d4.dtypes)
d4.head()
```

## Preparing Data

```python
#Read Diabetes.csv file
df = pd.read_csv('Diabetes.csv')
df.info()
```

```python
#See first 5 rows of a Dataset
df.head()
```

```python
#Fill empty values with Zero
df = df.fillna(0)
```

```python
#Move only specific columns to new dataset
df2 = df[['Glucose', 'BMI', 'Age', 'Outcome']]
```

```python
#See first 5 rows of a Dataset
df2.head()
```

```python
#See statistics information for numerical information
df2.describe()
```

```python
#will show if the condition (an entry) is zero the True else False
df2[df2.columns[:-1]] == 0
```

```python
#To drop the row if there is a Zero value (if the condition is not True (~), the pass it to df3)
df3 = df2.loc[~(df2[df2.columns[:-1]] == 0).any(axis=1)]
df3.describe()
df3.info()
```

```python
#Group the table by the column A Vs. all the other columns, and calculate the mean statistic
df3.groupby("Outcome").mean()
```

```python
#Group in a table but with different statistic calculations per column
df3.groupby("Outcome").agg({"Glucose":"mean", "BMI":"median", "Age":"sum"})
```

```python
#Group in a table but with different statistic calculations for all columns
df3.groupby("Outcome").agg(["mean", "median", "sum"])
```

```python
#Get the information of a column with specific values
positive = df3.loc[df3["Outcome"]==1]
negative = df3.loc[df3["Outcome"]==0]
print(positive.shape, negative.shape)
```

```python
#Saving file without the index column
df3.to_csv("clean_diabetes.csv", index=False)
```
