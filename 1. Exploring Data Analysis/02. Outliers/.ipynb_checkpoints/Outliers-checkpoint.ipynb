# Outliers

Some values must be removed, can be zeros, NaN values or extreme values that affect the analysis of our data. Let's see some strategies for this

```python
import numpy as np
import matplotlib.pyplot as plt

#We import three datasetws
d1 = np.loadtxt("outlier_1d.txt")
d2 = np.loadtxt("outlier_2d.txt")
d3 = np.loadtxt("outlier_curve.txt")
print(d1.shape, d2.shape)

#Then, we print the datasets d1 (blue) and d2 (multidimension-d2) where we can see some outliers
plt.scatter(d1, np.random.normal(7, 0.2, size=d1.size), s=1, alpha=0.5)
plt.scatter(d2[:, 0], d2[:, 1])
plt.show();
#Finally we print the d3 "outlier curve" filethat it is abouta dataset that have "huge outliers" for
#a normal analysis, and thus, we must analyze to remove them or not
plt.plot(d3[:, 0], d3[:, 1])
```

## Basics
1. Model the data as some analytics distribution
2. Find all points below a certain probability
3. Remove them
4. Refit the distributions, and potentially run again from Step 1.

```python
#Example with d1 (one dimension array values)
print("The total numbner of items in d1 are ", d1.shape)
```

```python
d1
```

```python
#First calculate the mean and standar deviation of our sample
mean, std = np.mean(d1), np.std(d1)
#Then we obtain anarray, of a score between the d1 values and the mean (distances between points of sample 
#and mean), then we divide them in the std deviation to see how far or near they are and then absolute
#to have a measurement of this distance, or in other words see the number of deviations for the points
z_score = np.abs((d1-mean)/std)
#We set a threshold (this is more intuitive than objective), change this value to test other thesholds
threshold = 3
#We save True in the array "good" if they are below the threshold or False if they are not
good = z_score < threshold

#We print how many items are nopt good (False), sum them and print (there were 5)
print(f"Rejection {(~good).sum()} points")
from scipy.stats import norm
#Calculation of threshold 3 is about x%percentage of the population (3 threshold to z_score items / 1010 entries) 
print(f"z_score of 3 corresponds to a prob of {100 * 2 * norm.sf(threshold):0.2f}%")
#For axis "Y", we create a random number of entries (this to visualize) of the same size of d1
visual_scatter = np.random.normal(size=d1.size)
#We print the values that are below the threshold in green and the above ones in red (5 False)
plt.scatter(d1[good], visual_scatter[good], s=2, label="Good", color="#4CAF50")
plt.scatter(d1[~good], visual_scatter[~good], s=2, label="Bad", color="#F44336")
plt.legend();
```

```python
from scipy.stats import multivariate_normal as mn

#We calculate the mean of d2 in all the vertical items (axis=0), and obtain the mean matrix, then
#we get the covariance matrix of the data with the Transpose of d2 (or else we have other calculation)
mean, cov = np.mean(d2, axis=0), np.cov(d2.T)
#We use the multivariate_normal to obtain the normal of the data, with the mean and covariance matrix
#then we take the Probability Density Function and check if this value is higher than 0.01/100 
#(0.00001), and put in the "good" array if it is True or False.
#In other words, it takes the d2 values, and shape them as a normal distribution, uses mean and 
#covariance to normalize the vriables of d2 and then compares it to a threshold of 0.01% (0.0001)
#Change the value of 0.01 to see different thresholds
good = mn(mean, cov).pdf(d2) > 0.01 / 100

#Finally, we plot d2 array, in X we use the first column of d2, where the values were True, in Y we
#do the same for the second column and print them as green.
#Then, we do the exact same but for the value s that are False as Bad
plt.scatter(d2[good, 0], d2[good, 1], s=2, label="Good", color="#4CAF50")
plt.scatter(d2[~good, 0], d2[~good, 1], s=2, label="Bad", color="#F44336")
plt.legend();
```

```python
#Save the d3 values (2d Dim Matix) into xs and ys
xs, ys = d3.T
# Then to get the values polynomial terms we use numpy's polyfit with a degree of 5th 
#(X, X^2, X^3, X^4, X^5)
p = np.polyfit(xs, ys, deg=5)
#This is to sample the Polynomial, giving the values of the polynomial terms (p) and the X (xs) variables
ps = np.polyval(p, xs)
#First we print the d3 values (xs and ys) as points with label "Data"
plt.plot(xs, ys, ".", label="Data")
#Then we print the polynomial curve obtained with ployfit of 5th degree vs xs
plt.plot(xs, ps, label="Bad poly fit")
plt.legend();

#As seen, we can see the outliers
```

Now, what we want to do is to give an score to each point to see how far is each one from the 
Polynomial-5th degree fit line, to have an idea how "near" or "far" are them and then set a 
threshold value to have this points set as Outliers.

```python
#Get a copy of the original data in case we want to re-run the code
x, y = xs.copy(), ys.copy()
#We will run this code for 5 times (1, 2, 3, 4, 5) as i (due that we re-use the 'x' and 'y' values
#this code will be done like an optimization iteration, removing the bad data in each code and 
#doing a similar analysis as the first one, but with the remaining "good" data)
for i in range(5):
    #Do a Polyfit of the data with degree 5th
    p = np.polyfit(x, y, deg = 5)
    #Get the 5th degree line with the p values and the X variables
    ps = np.polyval(p, x)
    #Set as True or false if array "y" minus array "ps" (calculated Plynomial)  is lower than 3
    good = y - ps < 3
    
    #Set new arrays as bad, where is False that this values are lower than 3
    x_bad, y_bad = x[~good], y[~good]
    #Set the arrays as "good", where is True that this values are lower than 3
    x, y = x[good], y[good]
    
    #Plot the "good" values
    plt.plot(x, y, ".", label="Used Data")
    #Print the 5th degree Polynomial line
    plt.plot(x, np.polyval(p,x), label=f"Poly fit {i}")
    #Plot the "bad" values
    plt.plot(x_bad, y_bad, ".", label="Not Used Data")
    plt.legend()
    plt.show()
    
    #If there are no "bad" values break the iteration
    if (~good).sum() == 0:
        break
```
