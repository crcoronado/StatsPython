 # Histogram 1D practical Examples
 *Example #1 and #2*

```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
```

```python
#Load data and get shapes
d1 = np.loadtxt("example_1.txt")
d2 = np.loadtxt("example_2.txt")
print(d1.shape, d2.shape)
```

## Histogram Plots

```python
#We plot both histograms, but the X axis is based only in the first histogram
plt.hist(d1, label="D1")
plt.hist(d2, label="D2")
plt.legend()
plt.ylabel("Counts");
```

```python
#To better plot these histograms, we get the bins from the min number of d1 and d2, to the maximum
#number of d1 and d2, with 50 groups. Now is easier to see both datasets
bins = np.linspace(min(d1.min(), d2.min()), max(d1.max(), d2.max()), 50)
counts1, _, _ = plt.hist(d2, bins=bins, label="D2")
plt.hist(d1, bins=bins, label="D1")
plt.legend()
plt.ylabel("Counts");
```

```python
#Changing Y axis from count to probability with density=True
bins = np.linspace(min(d1.min(), d2.min()), max(d1.max(), d2.max()), 50)
counts1, _, _ = plt.hist(d2, bins=bins, label="D2", density=True)
plt.hist(d1, bins=bins, label="D1", density=True)
plt.legend()
plt.ylabel("Probability");
```

```python
#Now the graphs are seen only with their border, lw is lineweight, ls is linestyle
bins = np.linspace(min(d1.min(), d2.min()), max(d1.max(), d2.max()), 50)
counts1, _, _ = plt.hist(d2, bins=bins, label="D2", density=True, histtype="step", lw=3)
plt.hist(d1, bins=bins, label="D1", density=True, histtype="step", ls=":")
plt.legend()
plt.ylabel("Probability");
```

```python
#More Tweeks
bins = np.linspace(min(d1.min(), d2.min()), max(d1.max(), d2.max()), 50)
plt.hist([d1, d2], bins=bins, label="Stacked", density=True, alpha=0.5, histtype="barstacked")
plt.hist(d1, bins=bins, label="D1", density=True, histtype="step", lw=1)
plt.hist(d2, bins=bins, label="D1", density=True, histtype="step", ls=":")
plt.legend()
plt.ylabel("Probability");
```

If **density=True** is used in all histograms in the same plot, then it is not needed to use the bin calculation with linspace
